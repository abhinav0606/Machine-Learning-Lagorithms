Importing the libraries

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Data preprocessing part as incliuding the feature scalling

```
dataset=pd.read_csv("/content/sample_data/Social_Network_Ads.csv")
dataset
```

```
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
X
Y
```

as not missing data and label encoding

```
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```
# applying the feature scalling part
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
```

```
X_test
```

```
# ssifcation class implimentation
from sklearn.svm import SVC
svc=SVC(kernel="linear",random_state=0)
svc.fit(X_train,Y_train)
svc.predict(X_test)
```

```
svc.predict(sc.transform([[30,45000]]))
```

```
y_pred=svc.predict(X_test)
z=np.concatenate((y_pred.reshape(len(y_pred),1),Y_test.reshape(len(Y_test),1)),1)
```

```
z
```

```
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(Y_test,y_pred)
```

```
accuracy_score(Y_test,y_pred)
```

```
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, svc.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
```
