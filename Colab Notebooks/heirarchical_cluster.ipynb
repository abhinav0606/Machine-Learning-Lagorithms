```
# importing the modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset=pd.read_csv("/content/sample_data/Mall_Customers.csv")
dataset
```

```
X=dataset.iloc[:,[3,4]].values
```

```
X
```

```
# creating a dendogram
import scipy.cluster.hierarchy as sh
dendogram=sh.dendrogram(sh.linkage(X,method="ward"))
plt.title("Dendogram Plot")
plt.xlabel("Clusters")
plt.ylabel("Euclidian Distance")
plt.show()
```

```
# training the dataset
from sklearn.cluster import AgglomerativeClustering
ag=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")
result=ag.fit_predict(X)
```

```
result
```

```
plt.scatter(X[result==0,0],X[result==0,1],c="red",label="Cluster1")
plt.scatter(X[result==1,0],X[result==1,1],c="blue",label="Cluster2")
plt.scatter(X[result==2,0],X[result==2,1],c="yellow",label="Cluster3")
plt.scatter(X[result==3,0],X[result==3,1],c="green",label="Cluster4")
plt.scatter(X[result==4,0],X[result==4,1],c="magenta",label="Cluster5")
plt.title("Plotting")
plt.xlabel("Annual Income")
plt.ylabel("Spending Hours")
plt.legend()
plt.show()
```
