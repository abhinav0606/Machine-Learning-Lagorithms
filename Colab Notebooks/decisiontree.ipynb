Coding the Decision Tree regression


Importing the libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
dataset=pd.read_csv("/content/sample_data/Position_Salaries.csv")
dataset
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values
```

```
X
```

```
Y
```

Now the data preprocessing is over as we dont have to do the feature scalling and as we have small dataset so we will not do the splitting of the dataset

```
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=0)
dt.fit(X,Y)
dt.predict([[6.5]])
```

```
plt.scatter(X,Y,color="red")
plt.plot(X,dt.predict(X),color="blue")
```

```
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape(len(X_grid),1)
plt.scatter(X,Y,color="red")
plt.plot(X_grid,dt.predict(X_grid),color="blue")
```
