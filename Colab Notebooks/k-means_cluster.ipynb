Importing the libraries for the module

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
dataset=pd.read_csv("/content/sample_data/Mall_Customers.csv")
dataset
```

```
# as we dont know the dependent variable so no y and not splitting into training and testing sets
X=dataset.iloc[:,3:5].values
X
```

```
X=dataset.iloc[:,[3,4]].values
```

```
X
```

```
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel("Number of cluster")
plt.ylabel("WCSS")
plt.title("WCSS Elbow Method")
print(wcss)
print(sum(wcss)/len(wcss))
```

```
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
y_kmeans=kmeans.fit_predict(X)
y_kmeans
```

```
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],c="red",label="c1")
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],c="blue",label="c2")
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],c="green",label="c3")
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],c="yellow",label="c4")
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],c="magenta",label="c5")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'cyan', label = 'Centroids')
```
