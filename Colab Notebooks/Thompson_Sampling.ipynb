```
# importing the modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# datapreprocessing
dataset=pd.read_csv("/content/sample_data/Ads_CTR_Optimisation.csv")
dataset
```

```
import random
d=10
N=10000
ads_selected=[]
ni_1=[0]*d
ni_0=[0]*d
total_rewards=0
for n in range(N):
  max_bound=0
  ad=0
  for i in range(d):
    theta=random.betavariate(ni_1[i]+1,ni_0[i]+1)
    if theta>max_bound:
      max_bound=theta
      ad=i
  reward=dataset.values[n,ad]
  ads_selected.append(ad)
  if reward==1:
    ni_1[ad]=ni_1[ad]+1
  else:
    ni_0[ad]=ni_0[ad]+1
  total_rewards=total_rewards+reward
```

```
dicty={}
for i in ads_selected:
  dicty[i]=ads_selected.count(i)
plt.bar(dicty.keys(),dicty.values())
```
